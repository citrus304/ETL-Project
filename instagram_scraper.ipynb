```python
#Importing instagram CSV
colnames = ['team', 'name', 'instagram']
data = pd.read_csv('instagram.csv', names=colnames, encoding = "utf_8")
data.head()
```

```python
#Importing unidecode
from unidecode import unidecode
```

```python
#Apply unidecode
data['name'] = data['name'].apply(unidecode)
```

```python
#Create the lists
team = data.team.tolist()
name = data.name.tolist()
instagram = data.instagram.tolist()
```

```python
#Create the empty lists
no_followers = []
no_following = []
```

```python
#Instagram Scraper

print("Beginning Data Retrieval")
print("------------------------")

for user in instagram:

    try:
        url = 'https://www.instagram.com/' + user
        r = requests.get(url).text

        start = '"edge_followed_by":{"count":'
        end = '},"followed_by_viewer"'
        followers= r[r.find(start)+len(start):r.rfind(end)]
        no_followers.append(followers)

        start = '"edge_follow":{"count":'
        end = '},"follows_viewer"'
        following= r[r.find(start)+len(start):r.rfind(end)]
        no_following.append(following)
        print (user, followers, following)
        
    except TypeError:
    
        print("*** Whoops. User not found! ***")
 
print("------------------------")
print("All done!")
    
    
```

```python
#Zipping the data
all_data = zip(team, name, instagram, no_followers, no_following)
```

```python
#Output file
output_file = os.path.join("instagram.csv")
```

```python
#Writing the rows
with open(output_file, "w") as datafile:
    writer = csv.writer(datafile)
    writer.writerow(["Team","Name","Instagram","No_followers", "No_following"])
    writer.writerows(all_data)
```
